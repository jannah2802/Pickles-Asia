```python
# Import libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

## Read & Analyze Data

```python
# Read DatiumTrain.rpt file
file_path = "Data/DatiumTrain.rpt"

# Parse tab-delimited data
parsed_data = []

with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
    # Read header
    header = f.readline().strip().split('\t')
    
    # Read data rows
    for line in f:
        row = line.strip().split('\t')
        parsed_data.append(row)

# Display first few rows
print("Header:", header[:5])  # Print first 5 column names
print("\nFirst 3 data rows:")
for i, row in enumerate(parsed_data[:3]):
    print(f"Row {i+1}: {row[:5]}")  # Print first 5 columns of each row

# Convert to DataFrame
DatiumTrain = pd.DataFrame(parsed_data, columns=header)
print(f"\nDataFrame shape for DatiumTrain: {DatiumTrain.shape}")
print(DatiumTrain.head())
```

## Data understanding 

### 1. Checking Missing Values

```python
# Calculate missing values per column
missing_counts = DatiumTrain.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]

if not missing_counts.empty:
    plt.figure(figsize=(16, 6))
    missing_counts.sort_values(ascending=False).plot(kind='bar')
    plt.title('Missing Values per Column in DatiumTrain')
    plt.ylabel('Number of Missing Values')
    plt.xlabel('Column Name')
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found in any column.")
```

```python
import numpy as np

# Replace string 'NULL' and 'null' with np.nan in the whole DataFrame
DatiumTrain.replace(['NULL', 'null'], np.nan, inplace=True)
DatiumTest.replace(['NULL', 'null'], np.nan, inplace=True)

# Now you can check for missing values again
missing_counts = DatiumTrain.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]

if not missing_counts.empty:
    plt.figure(figsize=(16, 6))
    missing_counts.sort_values(ascending=False).plot(kind='bar')
    plt.title('Missing Values per Column in DatiumTrain')
    plt.ylabel('Number of Missing Values')
    plt.xlabel('Column Name')
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found in any column.")
```
